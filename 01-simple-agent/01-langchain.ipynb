use [langchain](https://docs.langchain.com/oss/python/langchain/overview) to develop a simple agent

langchain v1.0 is available now!


before build an agent, wo need setup llm and mcp, they are core components for agent.


```python
# setup config
import os

base_url = os.getenv("BASE_URL", "DEFAULT")
api_key = os.getenv("API_KEY", "DEFAULT")
model = os.getenv("MODEL", "DEFAULT")
```

```python
# setup llm

from langchain_openai import ChatOpenAI


llm = ChatOpenAI(
    model=model,
    base_url=base_url,
    api_key=api_key,  # type: ignore
    extra_body={
        "thinking": {"type": "disabled"}
    },  # If your llm does not support thinking, you can ignore this param.
)
```

```python
# setup mcp

from langchain_mcp_adapters.client import MultiServerMCPClient

client = MultiServerMCPClient(
    {
        "finance-news": {
            "transport": "stdio",
            "command": "python",
            "args": ["../mcp_servers/finance_hot_news.py"],
        }
    }
)
```

```python
# build a simple agent, with tools
from langchain.agents import create_agent

tools = await client.get_tools()

agent = create_agent(
    model=llm,
    tools=tools,
    system_prompt="You are a financial analyst, you can use the tools to get the latest finance hot news.",
    debug=True,  # you can observe more details of the agent's execution
)

# run the agent
result = await agent.ainvoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "help me get the latest finance hot news, and analyze them, finally provice a markdown report",
            }
        ]
    }
)

# in the verbose `updates` info, you can find,
# step 1: accept human message and choose tools ->
# step 2: agent execute tool and get info ->
# step 3: agent execute llm and get reply
```
