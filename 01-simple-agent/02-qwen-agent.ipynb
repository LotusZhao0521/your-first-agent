use [qwen-agent](https://github.com/QwenLM/Qwen-Agent) to develop a simple agent.


```python
# setup config

import os

base_url = os.getenv("BASE_URL", "DEFAULT")
api_key = os.getenv("API_KEY", "DEFAULT")
model = os.getenv("MODEL", "DEFAULT")
```

```python
# setup llm

from qwen_agent.llm.oai import TextChatAtOAI

llm = TextChatAtOAI({"api_base": base_url, "api_key": api_key, "model": model})
```

```python
# build a simple agent, with tools

from qwen_agent.agents import Assistant

agent = Assistant(
    llm=llm,
    name="finance-analyst",
    description="A finance analyst agent",
    function_list=[
        {
            "mcpServers": {
                "finance-news": {
                    "command": "python",
                    "args": ["../mcp_servers/finance_hot_news.py"],
                }
            }
        }
    ],  # it seem more simple to bind mcp tools by config dict
)

messages = [
    {
        "role": "user",
        "content": "help me get the latest finance hot news, and analyze them, finally provice a markdown report",
    }
]
# run the agent
stream = agent.run(messages)  # it seems to have to stream output
for chunk in stream:
    last_state = chunk  # get the last response to get whole message
# ÁúüÁöÑÔºå‰ªñÂ±ÖÁÑ∂ÁúüÁöÑÊØèÊ¨°Â§öÂä†‰∏Ä‰∏™tokenËøîÂõûÔºåÊúÄÂêéËæìÂá∫Ë∑ü‰∏™ÈáëÂ≠óÂ°î‰∏ÄÊ†∑„ÄÇ
```
